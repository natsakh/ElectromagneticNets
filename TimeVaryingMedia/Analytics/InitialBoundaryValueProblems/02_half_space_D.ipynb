# Time-Switched Half-Space — Electric Displacement

## Problem formulation

A monochromatic plane wave propagates in an initially homogeneous, nondispersive dielectric medium with refractive index \(n_1\).

Before the temporal transition, the electric displacement is

$$
\mathbf{d}_0(z,t)
=
D_0e^{i\omega_1t}e^{-ik_1z}\,\mathbf{x},
$$

where

$$
D_0=\varepsilon_0n_1^2E_0,
\qquad
k_1=\frac{n_1\omega_1}{c}.
$$

At \(t=0\), the refractive index changes instantaneously only in the half-space \(z\ge0\):

$$
n(z)=
\begin{cases}
n_1, & z<0,\\
n_2, & z\ge0.
\end{cases}
$$

The region \(z<0\) remains unchanged, while the region \(z\ge0\) undergoes the temporal transition \(n_1\rightarrow n_2\).

## Wave equations

For \(t>0\), the electric displacement satisfies

$$
\frac{\partial^2d_1}{\partial z^2}
-
\frac{n_1^2}{c^2}
\frac{\partial^2d_1}{\partial t^2}
=
0,
\qquad z<0,
$$

and

$$
\frac{\partial^2d_2}{\partial z^2}
-
\frac{n_2^2}{c^2}
\frac{\partial^2d_2}{\partial t^2}
=
0,
\qquad z\ge0.
$$

## Laplace transform

We apply the Laplace transform with respect to time:

$$
D_j(z,p)
=
\mathcal{L}\{d_j(z,t)\}
=
\int_0^\infty d_j(z,t)e^{-pt}\,dt,
$$

with

$$
\mathcal{L}
\left\{
\frac{\partial^2d_j}{\partial t^2}
\right\}
=
p^2D_j(z,p)-p\,d_j(z,0)-\dot d_j(z,0).
$$

## Initial conditions

At an abrupt temporal interface, the electric displacement and its first time derivative are continuous.

Therefore, the same initial conditions hold in both spatial regions:

$$
d_j(z,0)
=
D_0e^{-ik_1z},
$$

$$
\dot d_j(z,0)
=
i\omega_1D_0e^{-ik_1z},
\qquad j=1,2.
$$

This is one of the main advantages of formulating the problem directly in terms of the electric displacement.

## Laplace-domain solution

The transformed electric displacement in the stationary half-space is

$$
D_1(z,p)
=
C_1^{(D)}(p)e^{n_1pz/c}
+
\frac{D_0}{p-i\omega_1}e^{-ik_1z},
\qquad z<0,
$$

whereas in the time-switched half-space,

$$
D_2(z,p)
=
C_2^{(D)}(p)e^{-n_2pz/c}
+
\frac{p+i\omega_1}{p^2+\omega_2^2}
D_0e^{-ik_1z},
\qquad z\ge0,
$$

with

$$
\omega_2=\frac{n_1}{n_2}\omega_1.
$$

The particular terms are the corresponding initial-value solutions. The terms containing \(C_1^{(D)}\) and \(C_2^{(D)}\) describe the additional fields generated by the spatial interface.

## Boundary conditions and coefficients

At the spatial interface, the tangential electric and magnetic fields are continuous. Since

$$
E_x=\frac{D_x}{\varepsilon_0n^2},
$$

the boundary conditions for the electric displacement are

$$
\frac{D_1(-0,p)}{n_1^2}
=
\frac{D_2(+0,p)}{n_2^2},
$$

and

$$
\frac{1}{n_1^2}
\left.
\frac{\partial D_1}{\partial z}
\right|_{z=-0}
=
\frac{1}{n_2^2}
\left.
\frac{\partial D_2}{\partial z}
\right|_{z=+0}.
$$

The resulting coefficients are related to the electric-field coefficients by

$$
C_1^{(D)}(p)
=
\varepsilon_0n_1^2C_1^{(E)}(p),
$$

$$
C_2^{(D)}(p)
=
\varepsilon_0n_2^2C_2^{(E)}(p).
$$

Consequently,

$$
D_1(z,p)
=
D_0\frac{n_1-n_2}{n_2}
\frac{p\,e^{n_1pz/c}}
{(p-i\omega_1)(p+i\omega_2)}
+
\frac{D_0e^{-ik_1z}}{p-i\omega_1},
\qquad z<0,
$$

and

$$
D_2(z,p)
=
D_0\frac{n_2-n_1}{n_1}
\frac{p(p+i\omega_1)e^{-n_2pz/c}}
{(p-i\omega_1)(p^2+\omega_2^2)}
+
D_0\frac{p+i\omega_1}{p^2+\omega_2^2}
e^{-ik_1z},
\qquad z\ge0.
$$

## Time-domain solution

Let \(\Theta(\cdot)\) denote the Heaviside step function.

For \(z<0\),

$$
\begin{aligned}
d_1(z,t)
={}&
D_0e^{i\omega_1t-ik_1z}
\\
&+
\Theta\left(t+\frac{n_1z}{c}\right)
D_0\frac{n_1-n_2}{n_1+n_2}
e^{i\omega_1\left(t+\frac{n_1z}{c}\right)}
\\
&+
\Theta\left(t+\frac{n_1z}{c}\right)
D_0\frac{n_1(n_1-n_2)}
{n_2(n_1+n_2)}
e^{-i\omega_2\left(t+\frac{n_1z}{c}\right)}.
\end{aligned}
$$

For \(z\ge0\),

$$
\begin{aligned}
d_2(z,t)
={}&
D_0\frac{n_1+n_2}{2n_1}
e^{i\omega_2t-ik_1z}
+
D_0\frac{n_1-n_2}{2n_1}
e^{-i\omega_2t-ik_1z}
\\
&+
\Theta\left(t-\frac{n_2z}{c}\right)
D_0\frac{2n_2^2}{n_1(n_1+n_2)}
e^{i\omega_1\left(t-\frac{n_2z}{c}\right)}
\\
&-
\Theta\left(t-\frac{n_2z}{c}\right)
D_0\frac{n_1+n_2}{2n_1}
e^{i\omega_2\left(t-\frac{n_2z}{c}\right)}
\\
&-
\Theta\left(t-\frac{n_2z}{c}\right)
D_0\frac{(n_1-n_2)^2}
{2n_1(n_1+n_2)}
e^{-i\omega_2\left(t-\frac{n_2z}{c}\right)}.
\end{aligned}
$$

The Heaviside factors describe the causal propagation of the disturbance generated at the spatial interface.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Python implementation

The following function evaluates the analytical solution for arbitrary \(n_1\), \(n_2\), \(\omega_1\), \(D_0\), and \(c\).

```python
def heaviside(x):
    return np.heaviside(x, 1.0)


def half_space_D_analytics(z, t, n1, n2, w1, D0=1.0, c=1.0):

    z = np.asarray(z)
    t = np.asarray(t)

    w2 = n1 / n2 * w1
    k1 = n1 * w1 / c

    tau_left = t + n1 * z / c
    tau_right = t - n2 * z / c

    d_left = (
        D0 * np.exp(1j * w1 * t - 1j * k1 * z)
        + heaviside(tau_left)
        * D0 * (n1 - n2) / (n1 + n2)
        * np.exp(1j * w1 * tau_left)
        + heaviside(tau_left)
        * D0 * n1 * (n1 - n2) / (n2 * (n1 + n2))
        * np.exp(-1j * w2 * tau_left)
    )

    d_right = (
        D0 * (n1 + n2) / (2 * n1)
        * np.exp(1j * w2 * t - 1j * k1 * z)
        + D0 * (n1 - n2) / (2 * n1)
        * np.exp(-1j * w2 * t - 1j * k1 * z)
        + heaviside(tau_right)
        * D0 * 2 * n2**2 / (n1 * (n1 + n2))
        * np.exp(1j * w1 * tau_right)
        - heaviside(tau_right)
        * D0 * (n1 + n2) / (2 * n1)
        * np.exp(1j * w2 * tau_right)
        - heaviside(tau_right)
        * D0 * (n1 - n2)**2 / (2 * n1 * (n1 + n2))
        * np.exp(-1j * w2 * tau_right)
    )

    return np.where(z < 0, d_left, d_right)
```

## Parameters and normalization

We use

$$
\omega_1=2\pi,
$$

so

$$
T_{\mathrm{ref}}
=
\frac{2\pi}{\omega_1}
=
1.
$$

The incident wavelength is

$$
\lambda_1
=
\frac{2\pi}{k_1}
=
\frac{2\pi c}{n_1\omega_1}.
$$

Time and space are displayed as \(t/T_{\mathrm{ref}}\) and \(z/\lambda_1\). The electric displacement is normalized by \(D_0\).

```python
n1 = 1.5
n2 = 1.0
w1 = 2 * np.pi
D0 = 1.0
c = 1.0

T_ref = 2 * np.pi / w1
k1 = n1 * w1 / c
lambda1 = 2 * np.pi / k1

print(f"T_ref   = {T_ref:.6f}")
print(f"lambda1 = {lambda1:.6f}")
```

## Temporal traces at fixed positions

```python
zeta_values = [-2, -1, 0, 1, 2]
t_norm = np.linspace(0, 5, 2000)
t = t_norm * T_ref

fig, axes = plt.subplots(len(zeta_values), 1, figsize=(10, 12), sharex=True)

for ax, zeta in zip(axes, zeta_values):

    z = zeta * lambda1
    d = half_space_D_analytics(z, t, n1, n2, w1, D0, c)

    ax.plot(t_norm, np.real(d) / D0, lw=2)
    ax.set_ylabel(r"$\mathrm{Re}\{d\}/D_0$")
    ax.set_title(rf"$z/\lambda_1={zeta}$")
    ax.grid(alpha=0.3)

axes[-1].set_xlabel(r"$t/T_{\mathrm{ref}}$")

plt.tight_layout()
plt.show()
```

## Space-time field map

```python
zeta = np.linspace(-2.5, 2.5, 700)
t_norm = np.linspace(0, 5, 700)

ZETA, T_NORM = np.meshgrid(zeta, t_norm)

Z = ZETA * lambda1
T = T_NORM * T_ref

D = half_space_D_analytics(Z, T, n1, n2, w1, D0, c)

plt.figure(figsize=(10, 6))

plt.pcolormesh(
    ZETA,
    T_NORM,
    np.real(D) / D0,
    shading="auto",
    cmap="RdBu_r"
)

plt.axvline(0, color="black", ls="--", lw=1.5)

plt.colorbar(label=r"$\mathrm{Re}\{d(z,t)\}/D_0$")
plt.xlabel(r"$z/\lambda_1$")
plt.ylabel(r"$t/T_{\mathrm{ref}}$")

plt.tight_layout()
plt.show()
```

## Spatial profiles at fixed times

```python
zeta = np.linspace(-2.5, 2.5, 1200)
z = zeta * lambda1

time_values = [0.0, 0.5, 1.0, 2.0]

plt.figure(figsize=(10, 6))

for t_norm_value in time_values:

    t_value = t_norm_value * T_ref
    d = half_space_D_analytics(z, t_value, n1, n2, w1, D0, c)

    plt.plot(
        zeta,
        np.real(d) / D0,
        lw=2,
        label=rf"$t/T_{{\mathrm{{ref}}}}={t_norm_value}$"
    )

plt.axvline(0, color="black", ls="--", lw=1.5)

plt.xlabel(r"$z/\lambda_1$")
plt.ylabel(r"$\mathrm{Re}\{d(z,t)\}/D_0$")

plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Two-dimensional field snapshot

Since the problem is one-dimensional, the field does not depend on the transverse coordinate. The two-dimensional snapshot is obtained by extending the spatial profile uniformly along a transverse direction.

```python
t_snapshot_norm = 1.0
t_snapshot = t_snapshot_norm * T_ref

zeta = np.linspace(-2.5, 2.5, 1000)
xeta = np.linspace(-0.6, 0.6, 250)

ZETA, XETA = np.meshgrid(zeta, xeta)

Z = ZETA * lambda1

D_snapshot = half_space_D_analytics(
    Z,
    t_snapshot,
    n1,
    n2,
    w1,
    D0,
    c
)

plt.figure(figsize=(11, 4))

plt.pcolormesh(
    ZETA,
    XETA,
    np.real(D_snapshot) / D0,
    shading="auto",
    cmap="RdBu_r"
)

plt.axvline(0, color="black", ls="--", lw=1.5)

plt.colorbar(label=r"$\mathrm{Re}\{d(z,t)\}/D_0$")

plt.xlabel(r"$z/\lambda_1$")
plt.ylabel(r"$x/\lambda_1$")

plt.title(
    rf"Electric-displacement snapshot at "
    rf"$t/T_{{\mathrm{{ref}}}}={t_snapshot_norm}$"
)

plt.tight_layout()
plt.show()
```
